# **Task 4: Bivariate Analysis**

| **Variable Type**          | **Best Graph Type**                        | **Example**                                      |
|---------------------------|-------------------------------------------|------------------------------------------------|
| **Numerical vs Numerical** | **Scatter Plot, Line Plot, Heatmap**      | Quantity vs Total Sales, Sales over Time       |
| **Categorical vs Numerical** | **Bar Plot, Box Plot, Violin Plot**    | Branch vs Total Sales, Payment Method vs Sales Distribution |
| **Categorical vs Categorical** | **Count Plot, Stacked Bar Chart**   | Customer Type vs Branch, Payment Method vs Gender |
| **Multiple Numerical Variables** | **Pair Plot, Heatmap**            | Correlation between Total, Tax, and Gross Income |

### **Graph Selection Guide:**
- **Scatter Plot:** Relationship between two numerical variables.
- **Bar Plot:** Compare numerical values across categories.
- **Box Plot / Violin Plot:** Show distribution and outliers for a numerical variable within categories.
- **Count Plot:** Compare frequency of categorical values.
- **Heatmap:** Show correlation between multiple numerical variables.
- **Pair Plot:** Explore multiple numerical relationships at once.


## Q1: Does the product line affect total sales? Which product line has the highest sales?

```python
# We want to check whether different product lines contribute differently to total sales.

sales_by_category = df.groupby('Product line')['Total'].sum().sort_values(ascending=False)
sales_by_category # pandas series
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=sales_by_category.index,y=sales_by_category.values,width=0.5)
plt.xticks(rotation=45)
plt.title('Total Sales by Product Line')
plt.ylabel('Total Sales')
```

```python
# Calculate Average Sales per Transaction, each row in the dataset represents a separate transaction or sale.

sales = df.groupby('Product line')['Total'].mean().sort_values(ascending=False)
sales
```

```python
plt.figure(figsize=(11,5))
sns.barplot(x=sales.index,y=sales.values)
plt.title('Average Sales per Transaction by Product Line')  
plt.ylabel('Average Sales') 
```

```python
# The data clearly shows that 'Food and Beverages' has the highest sales.
```

 # Q6: Is there a noticeable time trend in gross income?

### Montly Time Trend of Gross Income

```python
# we want to analyze trend on monthly basis so grouping the data by time(monthly).
df['Month'] = df['Date'].dt.to_period('M') # first extract month and year

# Group by Month and calculate the sum of Gross Income for each month
gross_income_by_month = df.groupby('Month')['gross income'].sum()
```

```python
sns.lineplot(x=gross_income_by_month.index.astype(str),y=gross_income_by_month.values)
plt.title('Montly Time Trend of Gross Income')
plt.ylabel('Gross income')
```

### Daily time trend of gross income

```python
plt.figure(figsize=(10,5))
gross_income_daily = df.groupby('Date')['gross income'].sum()
sns.lineplot(x=gross_income_daily.index,y=gross_income_daily.values)
plt.title('Daily Time Trend of Gross Income')
plt.ylabel('Gross income')
```

# Q7: Does the time of day affect sales? Are there peak hours or times when sales are higher?

```python
# extract hour from time, explicitly specify the format
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour
```

```python
sales_by_hour = df.groupby('Hour')['Total'].sum()
```

```python
sns.lineplot(x=sales_by_hour.index,y=sales_by_hour.values)
plt.title('Sales by Hour of Day')
plt.ylabel('Total Sales')
```

# Q8: Are different customer types (e.g., Member, Normal) concentrated in specific branches? How do customer types vary by branch?

```python
# size() - Counts total rows in each group.
df.groupby(['Customer type','Branch']).size() # returns a MultiIndex Series(the index is a combination of Branch and Customer type)
```

```python
# unstack(), it will take the second level of the index (Branch) and convert it into columns.
df.groupby(['Customer type','Branch']).size().unstack()
```

```python
data = df.groupby(['Customer type','Branch']).size().unstack()
data.plot(kind='bar',stacked=True)
plt.ylabel('Number of Customers')
plt.title('Customer Distribution by Branch and Customer Type')
```

# Q9: Are certain customer types more likely to purchase specific product lines? Is there a trend in buying behavior?

```python
dt = df.groupby(['Customer type','Product line']).size().unstack()
dt.plot(kind='bar',stacked=True)
plt.ylabel('Number of Purchases')
plt.xlabel('Product Line')
plt.title('Purchases per Customer Type and Product Line')
plt.legend(title='Customer Type', loc='upper left', bbox_to_anchor=(1,1)) 
```
